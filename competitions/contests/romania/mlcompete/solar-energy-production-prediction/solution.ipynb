# 2. Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.model_selection import cross_val_score,RandomizedSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

# 3. Data preparation

```python
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")
print(train.shape)
train.head(5)
```

```python
train.describe().round(3)
```

# 4. Models

```python
X = train.iloc[:,1:-1]
Y = train['energy_output']

pipeline = make_pipeline(StandardScaler(),GradientBoostingRegressor(random_state=0))
print("Score is:",cross_val_score(pipeline,X,Y,cv=5,scoring='neg_root_mean_squared_error').mean()*-1)
```

```python
pipeline.fit(X,Y)
predictions = pipeline.predict(test.iloc[:,1:])
```

# 5. Submission

```python
submission = pd.DataFrame({
    "SampleID": test['SampleID'],
    "energy_output": predictions,
})

submission.head()
```

```python
submission.to_csv("submission.csv", index=False)
```
